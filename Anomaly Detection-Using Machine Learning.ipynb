### Anomaly detection: 
    Use machine learning techniques to detect unusual patterns or outliers in a dataset.





Isolation forest is a machine learning algorithm for anomaly detection.
It's an unsupervised learning algorithm that identifies anomaly by isolating outliers in the data. Isolation Forest 
is based on the Decision Tree algorithm.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
# Prepare the data:
data = pd.read_csv("data.csv")
X = data.drop(['Outcome'], axis=1)
# plt.xticks(rotation=90)
```

```python
# Checking For null values
data.isnull().sum()
```

```python
data.boxplot()
plt.xticks(rotation=90)
plt.show()
```

```python
# Create an instance of IsolationForest:
clf = IsolationForest( max_samples=100, contamination='auto', n_jobs=-1)
```

```python
# Fit the model to the data:
clf.fit(X)
```

```python
# Use the model to predict the outliers:
outliers = clf.predict(X)
```

```python
# Identify the outliers:
outliers_index = np.where(outliers == -1)
```

```python
# Visualize the outliers:
import matplotlib.pyplot as plt
plt.scatter(X["Insulin"], X["Pregnancies"], color='black')
plt.scatter(X.iloc[outliers_index]["Insulin"], X.iloc[outliers_index]["Pregnancies"], color='red')
plt.xlabel("INSULIN")
plt.ylabel("PREGNANCIES")
plt.show()
```

```python
# Visualize the outliers:
import matplotlib.pyplot as plt
plt.scatter(X["Insulin"], X["Pregnancies"], color='black')
plt.scatter(X.iloc[outliers_index]["Insulin"], X.iloc[outliers_index]["Pregnancies"], color='red')
plt.xlabel("INSULIN")
plt.ylabel("PREGNANCIES")
plt.show()
```
